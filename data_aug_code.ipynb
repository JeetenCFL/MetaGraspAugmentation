```python
# defining transformations
D4_transforms = [
                A.HorizontalFlip(p=0.7),
                A.VerticalFlip(p=0.7),
                A.RandomRotate90(p=0.7),
                A.Transpose(p=0.7), 
                ]

Visual_transform = [
                    A.Blur (blur_limit=7, always_apply=False, p=0.2),
                    A.GaussianBlur (blur_limit=(3, 7), sigma_limit=0, always_apply=False, p=0.2),
                    #A.GaussNoise (var_limit=(10.0, 50.0), mean=0, per_channel=True, always_apply=False, p=0.2),
                    #A.ISONoise (color_shift=(0.01, 0.05), intensity=(1, 0.5), always_apply=False, p=0.2),
                    A.RandomBrightnessContrast (brightness_limit=0.2, contrast_limit=0.2, brightness_by_max=True, always_apply=False, p=0.4), 
                    ]
```

```python
# defining functions to convert bitmask back to polygon
import rasterio
from rasterio import features
import shapely
from shapely.geometry import Point, Polygon

def mask_to_polygons_layer(mask):
    all_polygons = []
    for shape, value in features.shapes(mask.astype(np.int16), mask=(mask >0), transform=rasterio.Affine(1.0, 0, 0, 0, 1.0, 0)):
        all_polygons.append(shapely.geometry.shape(shape))

    all_polygons = shapely.geometry.MultiPolygon(all_polygons)
    if not all_polygons.is_valid:
        all_polygons = all_polygons.buffer(0)
        # Sometimes buffer() converts a simple Multipolygon to just a Polygon,
        # need to keep it a Multi throughout
        if all_polygons.type == 'Polygon':
            all_polygons = shapely.geometry.MultiPolygon([all_polygons])
    return all_polygons
```

```python
def visualize_bbox(img, bbox, color=(255, 255, 0), thickness=2):  
    """Helper to add bboxes to images 
    Args:
        img : image as open-cv numpy array
        bbox : boxes as a list or numpy array in coco fromat [x_min, y_min, width, height]  
        color=(255, 255, 0): boxes color 
        thickness=2 : boxes line thickness
    """
    x_min, y_min, width, height = bbox
    x_min, y_min, width, height = int(x_min), int(y_min), int(width), int(height)
    cv2.rectangle(img, (x_min, y_min), (x_min+width, y_min+height), color=color, thickness=thickness)
    return img


def plot_image_anns(image, d_image, masks, boxes=None):
    """Helper to plot images with bboxes and masks 
    Args:
        image: image as open-cv numpy array
        d_image: depth image as open-cv numpy array
        masks: list of binary masks 
        bbox : boxes as a list or numpy array
    """
    # glue binary masks together
    one_mask = np.zeros_like(masks[0]).astype(np.uint8)
    for i, mask in enumerate(masks):
        one_mask += (mask > 0).astype(np.uint8) * (len(masks)-i) #
      
    if boxes is not None:
        for box in boxes:
            image = visualize_bbox(image, box)
            
    # for binary masks we get one channel and need to convert to RGB for visualization
    mask_rgb = label2rgb(one_mask, bg_label=0)            
        
    f, ax = plt.subplots(1, 3, figsize=(16, 16))             
    ax[0].imshow(image)
    ax[0].set_title('Original image')
    ax[1].imshow(d_image)
    ax[1].set_title('Original depth image')     
    ax[2].imshow(mask_rgb, interpolation='nearest')
    ax[2].set_title('Original mask')
    f.tight_layout()
    plt.show() 


def plot_image_aug(image, image_aug, d_image, aug_d_image, masks, aug_masks, bboxes, aug_bboxes):
    """Helper to plot images with bboxes and masks and their augmented versions 
    Args:
        image, image_aug: image as open-cv numpy array, original and augmented
        masks, aug_masks:setof binary masks, original and augmented 
        bbox, aug_boxes : boxes as a list or numpy array, original and augmented       
    """
    # glue masks together
    one_mask = np.zeros_like(masks[0]).astype(np.uint8)
    for i, mask in enumerate(masks):
        one_mask += (mask > 0).astype(np.uint8) * (len(masks)-i)
    
    one_aug_mask = np.zeros_like(aug_masks[0]).astype(np.uint8)
    for i, augmask in enumerate(aug_masks):
        one_aug_mask += (augmask > 0).astype(np.uint8) * (len(masks)-i)
    
    for box in bboxes:
        image = visualize_bbox(image, box)
    for augbox in aug_bboxes:
        image_aug = visualize_bbox(image_aug, augbox)    
        
    # for binary masks we get one channel and need to convert to RGB for visualization
    mask_rgb = label2rgb(one_mask, bg_label=0)            
    mask_aug_rgb = label2rgb(one_aug_mask, bg_label=0) 
    
    f, ax = plt.subplots(3, 2, figsize=(16, 16))             
    ax[0, 0].imshow(image)
    ax[0, 0].set_title('Original image')        
    ax[0, 1].imshow(image_aug)
    ax[0, 1].set_title('Augmented image')     
    ax[1, 0].imshow(mask_rgb, interpolation='nearest')
    ax[1, 0].set_title('Original mask')
    ax[1, 1].imshow(mask_aug_rgb, interpolation='nearest')
    ax[1, 1].set_title('Augmented mask')
    ax[2, 0].imshow(d_image)
    ax[2, 0].set_title('Original depth image')
    ax[2, 1].imshow(aug_d_image)
    ax[2, 1].set_title('Augmented depth image')
    f.tight_layout()
    plt.show() 
```

```python
new_img_index = 5237#This Id is not used till now for images
new_anno_index = 9168#This Id is not used till now for annotation
location = ""

for i in range(len(aug_req)):
    for j in range(aug_req[i][2]):
        # taking input from a list of the augmentable list
        imgID = random.choice(aug_req[i][1])
        row = df.loc[imgID]
        
        # I am only augmenting for classID aug_req[i][0] 
        catID = row['labels'].index(aug_req[i][0])
        
        dir = 'originalMetaGrasp/'
        image = cv2.imread(dir + row['imgPath'], cv2.IMREAD_UNCHANGED)
        d_image = cv2.imread(dir + row['dimgPath'], cv2.IMREAD_UNCHANGED)
        
        bitmasks = []
        poly = row['polygons'][catID]
        bitmasks.append(polygons_to_bitmask(poly, 1200, 1200))
        
        bboxes = [row['bboxes'][catID]]
        labels = np.array([row['labels'][catID]])
        layer = np.array([row['layer'][catID]])


        # # applying the transformation
        bbox_params={'format':'coco', 'label_fields': ['category_id']}
        augsD4 = A.Compose(D4_transforms, additional_targets={'d_image': 'image'}, bbox_params=bbox_params, p=1)       
        augmentedD4 = augsD4(image=image, d_image=d_image, masks=bitmasks, bboxes=bboxes, category_id=labels)     
        augsV = A.Compose(Visual_transform, p=1) 
        augmentedV = augsV(image=augmentedD4['image'])


        #getting the results back from the transformation
        aug_image = augmentedV['image']
        aug_d_image = augmentedD4['d_image']
        aug_masks = augmentedD4['masks']
        aug_bboxes = augmentedD4['bboxes']
        aug_area = []
        for mask in aug_masks:
            aug_area.append(mask.sum())


        #putting the mask back into the coco format
        aug_segmentation = []
        for k in range(len(aug_masks)):
            poly_i = list(mask_to_polygons_layer(aug_masks[k]).geoms)
            segmentation_aug_i = []
            for l,poly in enumerate(poly_i):
                x = enumerate(poly.exterior.coords.xy[0])
                segmentation_aug_i.append([])
                for m,xcor in x:
                    segmentation_aug_i[l].append(xcor)
                    segmentation_aug_i[l].append(poly.exterior.coords.xy[1][m])  
            aug_segmentation.append(segmentation_aug_i)

        
        #putting things in folder and json!!!
        img_dict = {
                  "file_name": row['imgPath'][:-9]+str(new_img_index)+'_rgb.png',
                  "height": 1200,
                  "width": 1200,
                  "id": new_img_index,
                  "difficulty": 1
                   }
        cv2.imwrite('/home/jeeten/MetaGrasp/originalMetaGrasp/'+img_dict['file_name'], aug_image, [cv2.IMWRITE_PNG_COMPRESSION,5])           
        cv2.imwrite('/home/jeeten/MetaGrasp/originalMetaGrasp/'+img_dict['file_name'][0:-8]+'_depth.png', aug_d_image, [cv2.IMWRITE_PNG_COMPRESSION,5])
        anno['images'].append(img_dict) 
        location = '/home/jeeten/'+img_dict['file_name']


        for index_num in range(labels.shape[0]):
            anno_dict = {
                "segmentation": aug_segmentation[index_num],
                "area": aug_area[index_num].tolist(),
                "iscrowd": 0,
                "image_id": new_img_index,
                "bbox": list(aug_bboxes[index_num]),
                "category_id": labels[index_num].tolist(),
                "id": new_anno_index,
                "layer": layer[index_num].tolist(),
                "occlusion": 0,
                "total_covered_edge": 0,
                "separated": False
            }

            anno['annotations'].append(anno_dict)
            new_anno_index = new_anno_index + 1
        new_img_index = new_img_index + 1  
        
    
out_file = open("post_aug_complete.json", "w")
json.dump(anno, out_file, indent = 2, cls=NumpyEncoder)
out_file.close()
```
