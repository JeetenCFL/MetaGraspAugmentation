```python
## for train image extraction
with open('/home/jeeten/MetaGrasp/train.json', 'r') as f:
    anno = json.load(f)

    # where is the meta-grasp original folder
    dir = '/home/jeeten/MetaGrasp/originalMetaGrasp/'
    targetDir = '/home/jeeten/MetaGrasp/MetaGraspV2/train/'

    for img in (anno['images']):
        rgbFile = dir+img['file_name']
        depthFile = dir+re.sub('rgb', 'depth', img['file_name'])
        newRGBName = targetDir+'images/'+str(img['id'])+'_rgb.png'
        newdepthName = targetDir+'depth_images/'+str(img['id'])+'_depth.png'
        
        shutil.move(rgbFile, newRGBName)
        shutil.move(depthFile, newdepthName)
```
